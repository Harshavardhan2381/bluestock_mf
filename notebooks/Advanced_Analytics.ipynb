## 5) Key Insights
1. **Historical VaR and CVaR analysis** shows that small-cap and thematic funds exhibit larger downside tail risk compared to diversified large-cap funds.
2. **Rolling 90-day Sharpe Ratios** varied significantly over time, indicating changing risk-adjusted performance across market cycles.
3. **Investor cohort analysis** reveals that the 2025 cohort invested a higher average SIP amount than the 2024 cohort, despite having a much smaller investor base.
4. **SIP continuity analysis** identified many investors with average transaction gaps exceeding 35 days, classifying them as potentially at-risk for SIP discontinuation.
5. **Sector concentration analysis (HHI)** indicates that most portfolios maintain moderate diversification, while a few funds show elevated concentration in specific sectors.
